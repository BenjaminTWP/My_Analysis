```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year_and_study_period = ["Y1LP1", "Y1LP2", "Y2LP1", "Y2LP2", "Y2LP3", "Y3LP1", "Y3LP2", "Y3LP3", "Y3LP4", "Y4LP1", "Y4LP2", "Y4LP3", "Y4LP4"]
reading_period_avg_hours = np.array([39.3, 38.7, 36.7, 39, 29.1, 28, 26.5, 37.9, 24.1, 41.8, 36.4, 27.4, 25]) 
reading_period_sum_grade = np.array([8, 8, 9, 10, 9, 10, 9, 8, 10, 7, 9, 9, 9])
```

```python
plt.scatter(reading_period_avg_hours, reading_period_sum_grade,
            color='magenta', linewidth=0.8, alpha=0.8)

coefs = np.polyfit(reading_period_avg_hours, reading_period_sum_grade, 1)
poly1d_fn = np.poly1d(coefs)

x_vals = np.linspace(min(reading_period_avg_hours), max(reading_period_avg_hours), 100)
plt.plot(x_vals, poly1d_fn(x_vals), color='black', linewidth=1)

for x, y, label in zip(reading_period_avg_hours, reading_period_sum_grade, year_and_study_period):
    plt.annotate(label, (x, y), xytext=(5, 3), textcoords='offset points', fontsize=8, rotation=30)

plt.xlabel('Average hours per study week')
plt.ylabel('Sum of grades for study period')
plt.title(f"Lean {round(coefs[0], 2)}, intersect {round(coefs[1], 2)}")
plt.show()
```
