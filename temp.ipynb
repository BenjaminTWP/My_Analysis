```python
import matplotlib.pyplot as plt

search = ['classification deep learning gnss signals jamming spoofing',
          'deep learning GNSS interference detection classification spoofing jamming',
          'machine learning GNSS signal anomaly detection spoofing jamming mitigation']
n_results = [[22, 23, 48, 75, 122, 206, 268, 348, 427, 571, 719],
             [19, 21, 41, 57, 102, 162, 223, 294, 370, 480, 634],
             [ 9, 20, 39, 41,  79, 120, 162, 229, 279, 409, 622],
             ]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

plt.figure(figsize=(10, 6))
plt.ylabel("Number of publications")
plt.title("Number of publications n (search results on google scholar) \n per year for a given search term")

for index in range(len(n_results)):
    plt.plot(year, n_results[index], label=search[index])
    
plt.xticks(year)
plt.legend()
plt.show()
```
